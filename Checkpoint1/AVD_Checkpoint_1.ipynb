## Bussiness Objective

Tujuan bisnis utama adalah meningkatkan performa akademik siswa berdasarkan faktor-faktor yang memengaruhinya. Sekolah ingin mengetahui variabel apa saja yang memiliki dampak signifikan terhadap nilai siswa agar dapat merancang kebijakan yang lebih efektif. Secara spesifik, tujuan bisnis meliputi:
1. Meningkatkan rata-rata nilai siswa.
2. Mengurangi jumlah siswa dengan performa rendah.
3. Mengoptimalkan program pendukung seperti bimbingan belajar dan konseling.
4. Membantu pengambilan keputusan berbasis data.

## Assess Situation

Sekolah memiliki data terkait berbagai faktor yang memengaruhi performa siswa, seperti jam belajar, kehadiran, dukungan orang tua, akses internet, dan motivasi belajar. Namun, data tersebut belum dimanfaatkan secara optimal untuk analisis mendalam.Tantangan yang dihadapi antara lain :
1. Belum diketahui faktor dominan yang paling berpengaruh.
2. Belum adanya sistem prediksi untuk mendeteksi siswa berisiko.
3. Pengambilan keputusan masih berbasis asumsi, bukan analisis data.

## Analytic Goals

Tujuan dari analitik dari proyek ini adalah :
1. Mengidentifikasi hubungan antara variabel independen dengan nilai akhir siswa.

3. Membangun model prediksi nilai atau kategori performa siswa (tinggi, sedang, rendah).
